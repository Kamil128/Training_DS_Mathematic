### Wariancja

W statystyce wariancję definiuje się jako miarę zmienności, która reprezentuje dystans członków grupy. Gdy wariancja zbioru danych jest mała, pokazuje bliskość punktów danych do średniej, podczas gdy większa wartość wariancji oznacza, że ​​obserwacje są bardzo rozproszone wokół średniej arytmetycznej i od siebie nawzajem.

### odchylenie standardowe
Odchylenie standardowe jest miarą, która określa ilościowo dyspersję obserwacji w zbiorze danych. Niskie odchylenie standardowe jest wskaźnikiem bliskości wyników do średniej arytmetycznej, wysokie odchylenie standardowe - wyniki są rozproszone w większym zakresie wartości.

```python
M
```

```python
M.std()
```

```python
M.std() ** 2 == M.var()
M.std() == np.sqrt(M.var())
```

### Kowariancaja, korelacja

Mówiąc prościej, oba pojęcia mierzą zależność i związek pomiędzy dwoma zmiennymi. “Kowariancja” = kierunek liniowej zależności między zmiennymi. Korelacja”, z drugiej strony, mierzy zarówno siłę, jak i kierunek liniowej zależności między dwiema zmiennymi. Korelacja jest funkcją kowariancji. To, co je wyróżnia, to fakt, że wartości korelacji są standaryzowane, a wartości kowariancji nie. Współczynnik korelacji dwóch zmiennych można uzyskać dzieląc kowariancję tych zmiennych przez iloczyn odchyleń standardowych tych samych wartości. Jeżeli definicja odchylenia standardowego zostanie zrewidowana, to zasadniczo mierzy ona bezwzględną zmienność rozkładu zbioru danych. Dzieląc wartości kowariancji przez odchylenie standardowe, zasadniczo skaluje się wartość w dół do ograniczonego zakresu od -1 do +1. Jest to dokładnie ten zakres wartości korelacji.

```python
# Wartości bez szumu
X = np.random.uniform(-10, 10, 10_000)
noise = np.random.randn(X.size)
y = X  + noise * 0

# X = np.sin(X)
# y = np.cos(y)

plt.figure(figsize=(7, 5))
plt.scatter(X, y, alpha=0.7, s=0.9)
plt.show()

print(f"Wariancja: {y.var()}")
print(f"Odchylenie standardowe: {y.std()}")
print(f"Koralacja:\n{np.corrcoef(X, y)}")  # [-1; 1]
print(f"Kowariancja:\n{np.cov(X, y)}")  # [0; +inf)
```

```python
# Mały szum
X = np.random.uniform(-10, 10, 10_000)
noise = np.random.randn(X.size)
y = X  + noise

# X = np.sin(X)
# y = np.cos(y)

plt.figure(figsize=(7, 5))
plt.scatter(X, y, alpha=0.7, s=0.9)
plt.show()

print(f"Wariancja: {y.var()}")
print(f"Odchylenie standardowe: {y.std()}")
print(f"Koralacja:\n{np.corrcoef(X, y)}")  # [-1; 1]
print(f"Kowariancja:\n{np.cov(X, y)}")  # [0; +inf)
```

```python
# Duży szum
X = np.random.uniform(-10, 10, 10_000)
noise = np.random.randn(X.size)
y = X  + noise * 10

# X = np.sin(X)
# y = np.cos(y)

plt.figure(figsize=(7, 5))
plt.scatter(X, y, alpha=0.7, s=0.9)
plt.show()

print(f"Wariancja: {y.var()}")
print(f"Odchylenie standardowe: {y.std()}")
print(f"Koralacja:\n{np.corrcoef(X, y)}")  # [-1; 1]
print(f"Kowariancja:\n{np.cov(X, y)}")  # [0; +inf)
```

```python
# Zależność nieliniowa
X = np.random.uniform(-10, 10, 10_000)
y = X

X = np.sin(X)
y = np.cos(y)

plt.figure(figsize=(7, 5))
plt.scatter(X, y, alpha=0.7, s=0.9)
plt.show()

print(f"Wariancja: {y.var()}")
print(f"Odchylenie standardowe: {y.std()}")
print(f"Koralacja:\n{np.corrcoef(X, y)}")  # [-1; 1]
print(f"Kowariancja:\n{np.cov(X, y)}")  # [0; +inf)
```
